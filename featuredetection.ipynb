# some feature detector & descriptor copilot list for:

## Feature Detectors:

- Fast Feature Detector (FAST): A corner detector used for real-time applications.
- Harris Corner Detector: A corner detector used for detecting corners in images.
- GFTT (Good Features to Track): A corner detector based on the Harris corner detection method.
- SIFT (Scale-Invariant Feature Transform): A method to detect and describe local features in images. It is patented and may not be available in all distributions of OpenCV.
- SURF (Speeded-Up Robust Features): An improved version of SIFT that is faster and more robust but still patented.
- ORB (Oriented FAST and Rotated BRIEF): A fast binary descriptor based on BRIEF, which is rotation invariant and robust to noise.
- BRISK (Binary Robust Invariant Scalable Keypoints): A fast binary descriptor that is scale and rotation invariant.
- MSER (Maximally Stable Extremal Regions): A method that selects stable regions in images based on intensity variations.
  \*StarDetector (CenSurE): A center-surround extrema feature detector.

## Feature Descriptors:

- SIFT (Scale-Invariant Feature Transform): SIFT can also be used as a feature descriptor in addition to being a feature detector.
- SURF (Speeded-Up Robust Features): Like SIFT, SURF can also be used as a feature descriptor.
- ORB (Oriented FAST and Rotated BRIEF): ORB can also be used as a feature descriptor.
- BRISK (Binary Robust Invariant Scalable Keypoints): BRISK can also be used as a feature descriptor.
- BRIEF (Binary Robust Independent Elementary Features): A lightweight descriptor that is faster to compute than SIFT or SURF.
- FREAK (Fast Retina Keypoint): A binary descriptor that is robust to scale changes.
- LATCH (Learned Arrangements of Three Patch Codes): A learned binary descriptor.
- DAISY: A descriptor suitable for dense wide baseline matching with high efficiency.
- VGG (Oxford Visual Geometry Group): A descriptor trained to improve performance on particular datasets.
- LUCID (Locally Uniform Comparison Image Descriptor): A descriptor designed for use with the CenSurE detector.
